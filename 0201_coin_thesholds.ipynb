```python
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt
```

```python
img = ski.data.coins()
```

```python
plt.subplots(figsize=(15, 10))
for i in range(12):
    plt.subplot(3, 4, i+1)
    threshold = i * 20 + 10
    img_i = img.copy()
    img_i[img_i < threshold] = 0
    img_i[img_i >= threshold] = 255
    plt.axis('off')
    plt.title(f"Threshold = {threshold}")
    plt.imshow(img_i, cmap="gray")

plt.show()
```
