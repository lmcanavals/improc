```python
import skimage as ski
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
r = ski.io.imread('ch03/polen.tif')
plt.imshow(r, cmap='gray')
```

```python
h = np.zeros((256))
n, m = r.shape
flatten = r.reshape(n*m)
for l in flatten:
    h[l] += 1
```

```python
plt.bar(list(range(256)), h)
```
