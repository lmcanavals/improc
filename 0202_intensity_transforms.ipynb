```python
import numpy as np
import skimage as ski
import matplotlib.pyplot as plt
```

```python
r = ski.data.camera()
L = 256
```

## Image Negatives

$$
s = L - 1 - r
$$

```python
plt.subplot(121)
plt.imshow(r, cmap='gray')
plt.subplot(122)
s = L - 1 - r
plt.imshow(s, cmap='gray')
```

## Log transformation $s = c \times log(1+r)$

```python
plt.subplots(figsize=(15, 9))
for i in range(6):
    c = i * 8 + 1
    plt.subplot(2, 3, i + 1)
    if i == 0:
        plt.imshow(r, cmap='gray')
        continue
    s = c * np.log(r) # L = 256 -> L - 1 = 255
    plt.axis('off')
    plt.title(f"c = {c}")
    plt.imshow(s, cmap='gray')

plt.show()
```

## Gamma Transformations $ s = c \times r ^ \gamma $

```python
gamma = [.04, .1, .2, .4, .67, 1, 1.5, 2.5, 5, 10, 25]
c = 1e-20
plt.subplots(figsize=(15, 9))
r = r.astype(float)
for i in range(12):
    plt.subplot(3, 4, i + 1)
    plt.axis('off')
    if i == 11:
        plt.imshow(r, cmap='gray')
        continue
    s =  c * r ** gamma[i]
    plt.title(f" = {gamma[i]}")
    plt.imshow(s, cmap='gray')

plt.show()
```
